#### 2.8 работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan

```python
# проверим, есть ли в нашем DataFrame пропущенные значения:
wine_df.isnull().sum()
# Увидим количество пропущенных значений в каждом 
```

```python
# Удаление строк с пропущенными значениями:
wine_df = wine_df.dropna(axis=0)
print("После удаления:")
wine_df
```

```python
# Намеренно сгенерируем пропущенные значения, прибивая какие-то куски данных в np.nan:
#  в этом примере я намеренно установил NaN в ячейки с индексами от 1 до 9 в столбцах с индексами от 1 до 4. 
import numpy as np
wine_df.iloc[1:10, 4:6] = np.nan
wine_df
```

```python
# Заполнение пропущенных значений средним значением по колонке:
# выбираем столбцы с числовыми значениями
numeric_cols = ['price', 'points']

# заполняем пропущенные значения средним по столбцу:
wine_df[numeric_cols] = wine_df[numeric_cols].fillna(wine_df[numeric_cols].mean())
print("Полсе заполнения:")
wine_df
```

#### 2.9. создание нового поля вычисленного на основе значений других полей: 
#### 2.9.1 через выражение на базе имеющихся колонок


```python
# создаем новый столбец "value_for_money"
wine_df['value_for_money'] = wine_df['points'] / wine_df['price']
wine_df
# Это выражение вычисляет отношение между очками, полученными за вино, и его ценой, 
# что может дать представление о том, насколько хорошо это вино соответствует своей стоимости.
```

#### 2.9.2 через DataFrame.apply

```python
# Создадим поле "points_ratio", которое будет содержать отношение баллов за вино к его цене
# Если цена равна нулю, то мы возвращаем NaN, чтобы показать, что значение неизвестно.
wine_df['points_ratio'] = wine_df.apply(lambda row: row['points'] / row['price'] if row['price'] > 0 else np.nan, axis=1)
wine_df
```

#### 2.9.3 через Series.apply


```python
# Cоздается новое поле price_category, которое определяется на основе значений столбца price следующим образом: 
# в случае, если значение price меньше 20, поле price_category получает значение "cheap", 
# если значение между 20 и 50, то "moderate", иначе "expensive".
wine_df['price_category'] = wine_df['price'].apply(lambda x: 'cheap' if x < 20 else 'moderate' if x < 50 else 'expensive')
wine_df
```
